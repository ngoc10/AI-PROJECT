#**XỬ LÝ ẢNH TRÙNG LẶP**

```python
import os
from PIL import Image
import hashlib

#Hàm này tính toán hash (MD5) của một hình ảnh. Hash là một giá trị duy nhất đại diện cho nội dung của hình ảnh.
def calculate_hash(image_path):
    """Calculate the hash of an image."""
    with Image.open(image_path) as img:
        hash = hashlib.md5(img.tobytes()).hexdigest()
    return hash

#Hàm này loại bỏ các hình ảnh trùng lặp trong một thư mục bằng cách so sánh hash của từng hình ảnh.
def remove_duplicate_images(folder_path):
    """Remove duplicate images in the specified folder."""
    image_hashes = {}
    duplicates = []

    for filename in os.listdir(folder_path):
        if filename.endswith(('.png', '.jpg', '.jpeg', '.bmp', '.gif')):
            file_path = os.path.join(folder_path, filename)
            image_hash = calculate_hash(file_path)

            if image_hash in image_hashes:
                duplicates.append(file_path)
            else:
                image_hashes[image_hash] = file_path

    for duplicate in duplicates:
        os.remove(duplicate)
        print(f"Removed duplicate image: {duplicate}")

# Đường dẫn tới thư mục chứa ảnh
folder_path = "/content/drive/MyDrive/Trí tuệ nhân tạo/Trí tuệ nhân tạo/Dataset/Image/Image Downloader"

# Gọi hàm để xóa ảnh trùng lặp
remove_duplicate_images(folder_path)
```

###Code sau đây:
- Đọc tất cả các file trong thư mục.
- Lọc ra các file hình ảnh.
- Sắp xếp danh sách file hình ảnh.
- Duyệt qua từng file hình ảnh, tạo tên mới theo thứ tự, và kiểm tra xem tên mới có trùng lặp không.
- Nếu tên mới không trùng lặp, đổi tên file; nếu có, thêm số đếm (counter) vào tên file mới cho đến khi không còn trùng lặp.
- Đổi tên file và in ra thông báo đã đổi tên.

#**ĐỔI TÊN ẢNH THEO STT 1-> ..**

```python
import os #Thư viện os cung cấp nhiều chức năng để tương tác với hệ điều hành, bao gồm việc làm việc với các file và thư mục.

def rename_files_in_folder(folder_path): #Hàm này nhận vào một tham số folder_path, là đường dẫn đến thư mục chứa các file hình ảnh cần đổi tên.
    files = os.listdir(folder_path)  #Lấy danh sách các file trong thư mục.

    image_files = [file for file in files if file.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp'))]
    image_files.sort()
    for i, filename in enumerate(image_files, start=1):  #Sử dụng enumerate để duyệt qua từng file trong image_files, với chỉ số bắt đầu từ 1 (start=1).
        file_extension = os.path.splitext(filename)[1]  #os.path.splitext(filename)[1] trả về phần mở rộng của file, ví dụ: .jpg.
        new_name = f"{i}{file_extension}"
        old_file = os.path.join(folder_path, filename)
        new_file = os.path.join(folder_path, new_name)
        #Nếu tên file mới đã tồn tại (os.path.exists(new_file)), thêm _counter vào tên file mới và tăng counter cho đến khi tìm được tên file mới không trùng lặp.
        counter = 1
        while os.path.exists(new_file):
            new_name = f"{i}_{counter}{file_extension}"
            new_file = os.path.join(folder_path, new_name)
            counter += 1
        os.rename(old_file, new_file)
        print(f"Đã đổi tên {old_file} thành {new_file}")

folder_path = '/content/drive/MyDrive/Trí tuệ nhân tạo/Trí tuệ nhân tạo/Dataset/Image/Image Downloader'
rename_files_in_folder(folder_path)
```

#**XUẤT FILE NAME RA EXCEL**

```python
import os
import pandas as pd

def list_image_files(folder_path):   #Lấy danh sách tất cả các tệp trong thư mục folder_path
    """List image files in the specified folder in numerical order."""
    image_files = sorted([f for f in os.listdir(folder_path) if f.endswith(('.png', '.jpg', '.jpeg', '.bmp', '.gif'))],
                         key=lambda x: int(os.path.splitext(x)[0]))   #Tách phần tên tệp (không bao gồm phần mở rộng) từ tên tệp đầy đủ x | #Sử dụng phần tên tệp dưới dạng số nguyên làm khóa để sắp xếp.
    return image_files

# Đường dẫn tới thư mục chứa ảnh trên Google Drive
folder_path = "/content/drive/MyDrive/Trí tuệ nhân tạo/Trí tuệ nhân tạo/Dataset/Image/Image Downloader"

# Gọi hàm để liệt kê các ảnh
image_files = list_image_files(folder_path)

# Tạo DataFrame từ danh sách ảnh
df = pd.DataFrame(image_files, columns=["Image Name"])

# Xuất DataFrame ra file Excel
output_path = "/content/drive/My Drive/image_files.xlsx"
df.to_excel(output_path, index=False)

print(f"Excel file saved to {output_path}")
```
